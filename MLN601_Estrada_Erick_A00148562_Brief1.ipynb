## 4.3. Build Model <a class="anchor"></a>
Run the modelling tool on the prepared dataset to create your model.

**Parameter settings** - With any modelling tool there are often a large number of parameters that can be adjusted. List the parameters and their chosen values, along with the rationale for the choice of parameter settings.

**Model** - This is the actual model produced by the modelling tool, not a report on the model.

***Parameter settings***

This is a manually implemented linear regression model with no hyperparameters. The slope (m) and intercept (c) were calculated directly using the least squares method. The model was chosen for its simplicity and ability to illustrate the linear relationship between two variables. The resulting equation, y = -0.03x + 0.56, yielded an R² value of 0.0637, indicating weak predictive power. To improve the model, future revisions could involve using multiple linear regression with additional predictors, applying log or polynomial transformations, or performing feature selection.

***Model***

```python
x = data['fixed acidity'].values
y = data['volatile acidity'].values

x_mean = np.mean(x)
y_mean = np.mean(y)

numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean)**2)
m = numerator / denominator
c = y_mean - m * x_mean

y_pred = m * x + c

r_squared = 1 - (np.sum((y - y_pred) ** 2) / np.sum((y - y_mean) ** 2))

print(f"Linear Regression Equation: y = {m:.2f}x + {c:.2f}")
print(f"R²: {r_squared:.4f}")

# Updated plot with new colors
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='darkcyan', label='Data Points')               # Changed to darkcyan
plt.plot(x, y_pred, color='crimson', label=f'Regression Line (R² = {r_squared:.2f})')  # Changed to crimson
plt.xlabel('Fixed Acidity')
plt.ylabel('Volatile Acidity')
plt.title('Linear Regression: Fixed Acidity vs Volatile Acidity')
plt.legend()
plt.grid(True)
plt.show()
```

***Model description***

The resulting model is a simple linear regression equation: y = -0.03x + 0.56, where y represents volatile acidity and x represents fixed acidity. The negative slope indicates a weak inverse relationship—meaning that as fixed acidity slightly increases, volatile acidity tends to decrease. The R² value of 0.0637 suggests that only about 6.4% of the variance in volatile acidity can be explained by fixed acidity, implying that this single predictor is not a strong determinant. The model is easy to interpret but limited in predictive power.
